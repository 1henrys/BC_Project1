```python
import pandas as pd
from pathlib import Path
import matplotlib
import hvplot.pandas
import plotly.express as px

%matplotlib inline
```

```python
# Original CSV has been modified and parsed from it's larger form. You can find the original data set here: https://www.dropbox.com/s/x4qmtfpwgns7176/weekly_housing_market_data_most_recent.csv?dl=0

file_path = Path("../data_imports/weekly_housing_market_data_most_recent.csv")

real_estate_index = pd.read_csv(file_path, encoding='utf-8')
```

```python
real_estate_index.drop(real_estate_index.iloc[:, 21:78], inplace=True, axis = 1)
real_estate_index.drop(real_estate_index.loc[:, 'duration':'percent_homes_sold_with_price_drops_yoy'].columns, inplace=True, axis = 1)
real_estate_index.drop(["region_id","region_type_id"], inplace=True, axis = 1)
```

```python
real_estate_index.dropna(inplace=True)
real_estate_index
```

```python
real_estate_index.set_index(["period_begin", "period_end"], inplace=True)
```

```python
real_estate_index.sort_values(by=["period_begin", "period_end", "region_name"])
```

```python
real_estate_index.to_csv(r'../data_imports/cleaned_weekly_housing_market_data.csv')
```
