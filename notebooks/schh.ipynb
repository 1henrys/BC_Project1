```python
import pandas as pd
import requests
import dotenv
import os
import json
from pathlib import Path
import numpy
%matplotlib inline
from datetime import datetime
import dateutil.parser
```

### Use the `load_dotenv()` method from the `dotenv` package to load and export the environment variables

```python
dotenv.load_dotenv()
```

### Use the `os.getenv` function to retrieve the environment variable named `QUANDL_API_KEY`. Store as a Python variable named `api_key`

```python
ms_api_key = os.getenv("MARKETSTACK_API_KEY")
ms_api_key_string = "&access_key=" + ms_api_key 
ms_api_key_string
```

### Concatenate `request_url` with the `api_key` variable

```python
dotenv.load_dotenv(dotenv.find_dotenv())
```

```python
#ms_request_url = "https://api.marketstack.com/v1/exchanges/INDX/tickers?"
ms_request_url = "https://api.marketstack.com/v1/eod?symbols=SCHH"
ms_request_url = ms_request_url + ms_api_key_string + "&limit=3000" #&sort=DESC&date_from=YYYY-MM-DD&date_to=YYYY-MM-DD&offset=0
ms_request_url
```

### Execute GET request with API key

```python
# Execute get request
ms_response_data = requests.get(ms_request_url)
```

### Display content to screen using the content attribute

```python
# Output data
ms_data = ms_response_data.json()
#print(json.dumps(data, indent=4))
ms_list = ms_data["data"]
ms_list
```

```python
column_list = ['date', 'schh_adj_closing']
ms_df = pd.DataFrame(columns=column_list)

for dict in ms_df: 
    for list in dict: 
        if list == "date":
            temp_date = str(dateutil.parser.parse(dict[list]).date())
            #print(temp_date)
        elif list == "adj_close":
            temp_close = dict[list] 
            
    new_row = {'date':temp_date, 'vnq_adj_closing':temp_close}
    ms_df = ms_df.append(new_row, ignore_index=True)                 
    
    
print(ms_df)
```
